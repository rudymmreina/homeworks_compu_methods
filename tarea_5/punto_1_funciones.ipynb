```python
%pylab inline
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq, fft2
```

```python
Im3 = pylab.imread("ves_full_150_002-1.tif")
fig = figure(1, figsize=(10.5,8.5))
imshow(Im3[:,:], cmap = cm.Greys_r)
plt.title('Imagen Original', fontsize=12)
```

```python
#print Im1
#haciendo la transformada de fourier discreta 2D
imagen_fourier=np.fft.fft2(Im2)
print imagen_fourier[0,0]
imagen_fourier[0,0]=0
imagen_fourier_shift=np.fft.fftshift(imagen_fourier)
magnitud_espectro_imagen=np.abs(imagen_fourier_shift)

plt.imshow(magnitud_espectro_imagen, cmap = cm.Greys_r)
plt.title("Espectro completo")
plt.show()
fig = figure(1, figsize=(10.5,10.5))
plt.imshow(magnitud_espectro_imagen[90:160,140:220]  , cmap = cm.Greys_r)
plt.title("Espectro aumentado")
```

```python
rows,cols=Im2.shape
mask_cir_low_smooth=np.zeros((rows,cols),float)

crow,ccol=rows/2 , cols/2
f_off=12
w=5
#print mask_cir[1,2]
for i in range(rows):
    for j in range(cols):
        d=math.sqrt((crow-i)**2+(ccol-j)**2)
        if d<(f_off-w):
            mask_cir_low_smooth[i,j]=1
        elif d>(f_off+w):
            mask_cir_low_smooth[i,j]=0
        else:
            mask_cir_low_smooth[i,j]=0.5*(1-(math.sin(pi*(d-f_off)/(2*w))))
plt.imshow(mask_cir_low_smooth, cmap = cm.Greys_r)
plt.title("mascara circular low pass filter")
plt.show()
```

```python
rows,cols=Im2.shape
mask_cir_high_smooth=np.zeros((rows,cols),float)

crow,ccol=rows/2 , cols/2
f_off=12 #frecuencia de corte
w=5#frecuencia de suavizado
#print mask_cir[1,2]
for i in range(rows):
    for j in range(cols):
        d=math.sqrt((crow-i)**2+(ccol-j)**2)
        if d<(f_off-w):
            mask_cir_high_smooth[i,j]=0.0
        elif d>(f_off+w):
            mask_cir_high_smooth[i,j]=1.0
        else:
            mask_cir_high_smooth[i,j]=((math.sin(pi*(d-f_off)/(2*w))))
#plt.imshow(mask_cir_high_smooth, cmap = cm.Greys_r)
#plt.title("mascara circular low pass filter")
#plt.show()


mask_cir_high2=np.zeros((rows,cols),np.uint8)
crow,ccol=rows/2 , cols/2
#print mask_cir[1,2]
for i in range(rows):
    for j in range(cols):
        d=math.sqrt((crow-i)**2+(ccol-j)**2)
        if d<(f_off-w):
            mask_cir_high2[i,j]=0
        else:
            mask_cir_high2[i,j]=1
#plt.imshow(mask_cir_high2, cmap = cm.Greys_r)
#plt.title("mascara circular high pass filter")
#plt.show()

mask_smooth_high=mask_cir_high_smooth+mask_cir_high2

plt.imshow(mask_smooth_high, cmap = cm.Greys_r)
plt.title("mascara circular high pass filter")
plt.show()
```

```python
imagen2_fourier=np.fft.fft2(Im2)
#print imagen2_fourier[0,0]
imagen2_fourier[0,0]=0
img2_four_shift=np.fft.fftshift(imagen2_fourier)
mag_espec_imagen2=np.abs(img2_four_shift)


rows,cols=Im2.shape
crow,ccol=rows/2 , cols/2

#apply mask low and inverse DFT
frecuencia_filter_low2=mag_espec_imagen2*mask_cir_low_smooth
frecuencia_imag_low2=img2_four_shift*mask_cir_low_smooth


imagen2_ishift=np.fft.ifftshift(frecuencia_imag_low2)
imagen2_back_low=np.fft.ifft2(imagen2_ishift)
imagen2_low=np.abs(imagen2_back_low)

#apply mask high and inverse DFT

frecuencia_filter_high2=mag_espec_imagen2
frecuencia_imag_high2=img2_four_shift*mask_smooth_high


imagen2_ishift_high=np.fft.ifftshift(frecuencia_imag_high2)
imagen2_back_high=np.fft.ifft2(imagen2_ishift_high)
imagen2_high=np.abs(imagen2_back_high)


plt.imshow(Im2, cmap = cm.Greys_r)
plt.title("original")
plt.show()

plt.imshow(imagen2_low, cmap = cm.Greys_r)
plt.title("low pass filtering")
plt.show()

plt.imshow(imagen2_high, cmap = cm.Greys_r)
plt.title("high pass filtering ")
plt.show()

```
