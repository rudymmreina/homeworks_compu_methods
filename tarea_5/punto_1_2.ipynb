```python
%pylab inline
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq, fft2
```

# Definicion de funciones

```python

#funcion que encuentra la transformada de fourier discreta 2D y muestra el espectro de frecuencias
def fourier_2d(imagen):
    im_fft=np.fft.fft2(imagen)
    im_fft[0,0]=0
    im_sfft=np.fft.fftshift(im_fft)
    im_sfft_mag=np.abs(im_sfft)
    return im_sfft_mag
```

```python
#funcion que crea una mascara circular de la imagen suave pasa bajas frecuencias
#parametros de entrada
#     imagen=matriz, arreglo dimensional
#     f= frecuencia de corte 
#     W= frecuencia de suavizado
#parametros de salida
    #mask_cir_low_smooth=matriz que muestra la mascara
def low_pass_filter_smooth(imagen,f,W):
    rows,cols=imagen.shape
    mask_cir_low_smooth=np.zeros((rows,cols),float)

    crow,ccol=rows/2 , cols/2
    f_off=f
    w=W
    #print mask_cir[1,2]
    for i in range(rows):
        for j in range(cols):
            d=math.sqrt((crow-i)**2+(ccol-j)**2)
            if d<(f_off-w):
                mask_cir_low_smooth[i,j]=1
            elif d>(f_off+w):
                mask_cir_low_smooth[i,j]=0
            else:
                mask_cir_low_smooth[i,j]=0.5*(1-(math.sin(pi*(d-f_off)/(2*w))))
    return mask_cir_low_smooth

```

```python
#funcion que crea una mascara circular de la imagen suave pasa altas frecuencias
#parametros de entrada
#     imagen=matriz, arreglo dimensional
#     f= frecuencia de corte 
#     W= frecuencia de suavizado
#parametros de salida
    #mask_cir_high_smooth=matriz que muestra la mascara
def high_pass_filter_smooth(imagen,f,W):
    rows,cols=imagen.shape
    mask_cir_high_smooth=np.zeros((rows,cols),float)

    crow,ccol=rows/2 , cols/2
    f_off=f #frecuencia de corte
    w=W#frecuencia de suavizado
    
    for i in range(rows):
        for j in range(cols):
            d=math.sqrt((crow-i)**2+(ccol-j)**2)
            if d<(f_off-w):
                mask_cir_high_smooth[i,j]=0.0
            elif d>(f_off+w):
                mask_cir_high_smooth[i,j]=1.0
            else:
                mask_cir_high_smooth[i,j]=((math.sin(pi*(d-f_off)/(2*w))))



    mask_cir_high2=np.zeros((rows,cols),np.uint8)
   
    for i in range(rows):
        for j in range(cols):
            d=math.sqrt((crow-i)**2+(ccol-j)**2)
            if d<(f_off-w):
                mask_cir_high2[i,j]=0
            else:
                mask_cir_high2[i,j]=1


    mask_smooth_high=mask_cir_high_smooth+mask_cir_high2
    return mask_smooth_high

```

```python
def filtro(imagen,mascara):

    imagen2_fourier=np.fft.fft2(imagen)
    imagen2_fourier[0,0]=0
    img2_four_shift=np.fft.fftshift(imagen2_fourier)
   
    #apply mask  and inverse DFT
    
    frecuencia_imag_low2=img2_four_shift*mascara
    imagen2_ishift=np.fft.ifftshift(frecuencia_imag_low2)
    imagen2_back_low=np.fft.ifft2(imagen2_ishift)
    imagen_filtrada=np.abs(imagen2_back_low)

    return imagen_filtrada



```

```python
#cargando la imagen, insertar el nombre de la imagen
im=pylab.imread("colesterol-1.tif")
plt.imshow(im, cmap = cm.Greys_r)
plt.title("Imagen original")
plt.show()
```

```python
#imprimiendo el espectro de frecuencias
frecuencias=fourier_2d(im)

plt.imshow(frecuencias, cmap = cm.Greys_r)
plt.title("Espectro de frecuencias completo")
plt.show()
fig = figure(1, figsize=(10.5,10.5))
plt.imshow(frecuencias[90:160,140:220]  , cmap = cm.Greys_r)
plt.title("Espectro de frecuencias aumentado")
```

```python
#imprimiendo las mascaras alta y baja
mask_low=low_pass_filter_smooth(im,12,5)
plt.imshow(mask_low, cmap = cm.Greys_r)
plt.title("mascara circular low pass filter")
plt.show()
mask_high=high_pass_filter_smooth(im,12,5)
plt.imshow(mask_high, cmap = cm.Greys_r)
plt.title("mascara circular high pass filter")
plt.show()
```

```python
#aplicando los filtros
imagen_filtrada_alto=filtro(im,mask_high)
imagen_filtrada_bajo=filtro(im,mask_low)


plt.imshow(imagen_filtrada_bajo, cmap = cm.Greys_r)
plt.title("low pass filtering greys")
plt.show()

plt.imshow(imagen_filtrada_alto, cmap = cm.Greys_r)
plt.title("high pass filtering ")
plt.show()
```
