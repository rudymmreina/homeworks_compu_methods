```python
%pylab inline
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq, fft2
```

```python
circulos = loadtxt("circulos.dat")
x = circulos[:,0]
y = circulos[:,1]
scatter(x,y,s=0.1)
```

```python
radio=[]
for i in range(29010):
    r=math.sqrt((circulos[i,0])**2+(circulos[i,1])**2)
    radio.append(r)
```

```python
#funcion de correlacion 
def correlacion(l):
    rho=[]
    for i in range(29010-l):
        p=radio[i]*radio[i+l]
        rho.append(p)
    Rho=sum(rho)
    return Rho
```

```python
func_corr=[]
for j in range(29010):
    c=correlacion(j)
    func_corr.append(c)
```

```python
plot(func_corr)
```

```python
fft_r = fft(func_corr)
```
